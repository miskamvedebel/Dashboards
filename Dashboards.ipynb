```python
import pandas as pd
import numpy as np
from IPython.display import HTML
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
data_female = pd.read_excel('Indicator_TC female ASM.xlsx')
data_female = data_female.rename({'TC female (mmol/L), age standardized mean': 'country'}, axis=1)
data_female = pd.melt(data_female, id_vars=['country'], var_name='year', value_name='level')
data_male = pd.read_excel('Indicator_TC male ASM.xlsx')
data_male = data_male.rename({'TC male (mmol/L), age standardized mean': 'country'}, axis=1)
data_male = pd.melt(data_male, id_vars=['country'], var_name='year', value_name='level')
fem_2008= data_female[data_female['year'] == 2008].reset_index(drop=True)
male_2008= data_male[data_male['year'] == 2008].reset_index(drop=True)
all_2008 = pd.concat([fem_2008, male_2008], axis=0).reset_index(drop=True)
```

```python
all_2008 = all_2008.groupby(['country', 'year'])['level'].mean().reset_index()
```

```python
data_male = dict(
                type='choropleth',
                autocolorscale = False,
                locations = male_2008['country'],
                z = male_2008['level'],
                locationmode = 'country names'
               )
data_female = dict(
                type='choropleth',
                autocolorscale = False,
                locations = fem_2008['country'],
                z = fem_2008['level'],
                locationmode = 'country names'
               )
all_ = dict(
                type='choropleth',
                autocolorscale = False,
                locations = all_2008['country'],
                z = all_2008['level'],
                locationmode = 'country names'
               )
data = [data_male, data_female, all_]

updatemenus = list([
    dict(
        x = 0.1,
        xanchor = 'left',
        y = 1.0,
        yanchor = 'top'),
    dict(buttons=list(
        [
            dict(args = ["visible", [True, False, False]],
                 label = 'Males',
                 method = 'restyle'),
            dict(args = ["visible", [False, True, False]],
                 label = 'Females',
                 method = 'restyle'),
            dict(args = ["visible", [False, False, True]],
                 label = 'All',
                 method = 'restyle')
        ]),
    )
])


layout = dict(
            title = 'Cholesterol level mmol/L 2008', 
            updatemenus=updatemenus, margin = dict( t=30, b=30, l=0, r=0 )
            )

fig = dict(data=data, layout=layout)
iplot(fig)
```
